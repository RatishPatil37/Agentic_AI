## **C. Document Structure Based Splitting**

### What is it?
- >This strategy leverages the inherent structure of documents (HTML
tags, Markdown headers, JSON objects, code functions) to create
meaningful chunks. Each chunk respects the logical organization of the
document.
- >It's like organizing a book by chapters and sections rather than just
cutting it every 1000 words.

> Advantages:
- Preserves logical document organization
- Maintains semantic meaning within chunks
- Metadata-rich chunks (headers, section info)
- Excellent for structured documents
- Better retrieval accuracy
- Natural chunk boundaries

> Disadvantages:
- Only works with structured documents
- Requires document-specific splitters
- May create very large chunks if sections are long
- More complex to implement
- Not suitable for unstructured text
- May need combination with other strategies

````python
from langchain_text_splitters import RecursiveCharacterTextSplitter, Language
from pprint import pp
from termcolor import COLORS, colored
from random import choice
````

````python
python_code= """
import numpy as np
from typing import List, Optional

def calculate_mean(numbers: List[float]) -> float:
    '''Calculate the arithmetic mean of a list of numbers.

    Args:
    numbers: List of numerical values

    Returns:
    The mean value
    '''
    return sum(numbers) / len(numbers)

def calculate_median(numbers: List[float]) -> float:
    '''Calculate the median of a list of numbers.'''
    sorted_nums = sorted(numbers)
    n = len(sorted_nums)
    mid = n // 2

    if n % 2 == 0:
        return (sorted_nums[mid - 1] + sorted_nums[mid]) / 2
    
    return sorted_nums[mid]

class StatisticalAnalyzer:
    '''A class for performing statistical analysis on datasets.'''

    def _init_(self, data: List[float]):
        self.data = data
        self.mean = None
        self.median = None

    def analyze(self) -> dict:
        '''Perform complete statistical analysis.'''
        self.mean = calculate_mean(self.data)
        self.median = calculate_median(self.data)

        return {
        'mean': self.mean,
        'median': self.median,
        'count': len(self.data)
        }

def get_summary(self) -> str:
    '''Return a formatted summary of the analysis.'''
    if self.mean is None:
        self.analyze()

    return f"Mean: {self.mean :. 2f}, Median: {self.median :. 2f}"

def main():
    '''Main execution function.'''
    data = [1.5, 2.3, 3.7, 4.2, 5.1]
    analyzer = StatisticalAnalyzer(data)
    results = analyzer.analyze()
    print(analyzer.get_summary())

if __name__ == "__main__":
    main()
"""
````

````python
py_splitter= RecursiveCharacterTextSplitter.from_language(
    language= Language.PYTHON,
    chunk_size= 600,
    chunk_overlap= 50 
)

code_chunks= py_splitter.split_text(python_code)
[chunk for chunk in code_chunks]
````

If you want to see the overlap in action, you need to force the splitter to make "bad" cuts. Try lowering your chunk_size to something very small, like 150, and keep the chunk_overlap at 55.

Because a single function won't fit into 150 characters, the splitter will be forced to slice the functions in half, and you will immediately see 55 characters of overlap appear between the chunks!

````python
def display_chunks(chunks: list[str], chunk_overlap):
    colors_list= list(COLORS.keys())[2:8]
    print(f"Number of chunks: {len(chunks)}\n")

    for num, chunk in enumerate(chunks, 1):
        print(f"Chunk {num}: Length of characters: {len(chunk)}")
        print(colored(chunk, choice(colors_list)),"\n")

display_chunks(code_chunks, 55)
````

### **Markdown Splitters**

````python
from langchain_text_splitters import MarkdownHeaderTextSplitter

headers_to_split_on= [
    ("#", "Header_1"),
    ("##", "Header_2"),
    ("###", "Header_3")
]

md_head_splitter= MarkdownHeaderTextSplitter(
    headers_to_split_on= headers_to_split_on,
    strip_headers= False
)

md_head_chunks= md_head_splitter.split_text(markdown)

print(f"Number of Chunks: {len(md_head_chunks)}\n")
# pp(md_head_chunks)

for num, doc in enumerate(md_head_chunks, 1):
    print(f"Chunk {num}: \n{doc.page_content}", end="\n\n")
````
